```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
plt.rcParams['figure.figsize']= [15,8]
```

```python
sns.regplot(x=hdb_df['floor_area_sqm'],y=hdb_df['resale_price'])
plt.show()
```

```python
sns.boxplot(x=hdb_df['town'],y=hdb_df['resale_price'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x=hdb_df['lng'], y=hdb_df['lat'], hue=hdb_df['resale_price'])
plt.show()
```

```python
sns.boxplot(x=hdb_df['town'],y=hdb_df['resale_price']/hdb_df['floor_area_sqm']/hdb_df['remaining_lease'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x=hdb_df['lng'], y=hdb_df['lat'], hue=hdb_df['resale_price']/hdb_df['floor_area_sqm']/hdb_df['remaining_lease'],
               )
plt.title('Resale Prices psqm plyr in Singapore')
plt.show()
```

```python
sns.boxplot(x=hdb_df['flat_type'],y=hdb_df['resale_price'],
           order=['1 ROOM','2 ROOM','3 ROOM', '4 ROOM', '5 ROOM', 'EXECUTIVE'])
plt.xticks(rotation=90)
plt.title('Resale Prices Against Flat Type')
plt.show()
```

```python
sns.boxplot(x=hdb_df['flat_type'],y=hdb_df['resale_price']/hdb_df['floor_area_sqm'], 
            order=['1 ROOM','2 ROOM','3 ROOM', '4 ROOM', '5 ROOM', 'EXECUTIVE'])
plt.xticks(rotation=90)
plt.title('Resale Prices sqm Against Flat Type')
plt.show()
```
